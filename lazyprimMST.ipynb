to create a visited list with length of total nodes (here total of 8)

```python
def create_visited():
    visited = [False] * number_of_nodes
    visited[0] = True
    return visited
```

creating a dictionary with key length of total nodes (here total of 8), we use this to store our new minimum spanning tree

```python
def create_mst_dict():
    mst = {}
    for i in range(0,number_of_nodes):
        mst[i] = []
    return mst
```

of mst

```python
def re_create_path(origin,to,weight):
    global node_counter
    global path_cost
    mst[origin].append(to)
    node_counter += 1
    path_cost += weight
```
